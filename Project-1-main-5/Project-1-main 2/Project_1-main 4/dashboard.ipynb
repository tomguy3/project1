```python
import pandas as pd
from pathlib import Path
import panel as pn
from panel.interact import interact
pn.extension("plotly")
import plotly.express as px
import hvplot.pandas
import matplotlib.pyplot as plt
import os
import numpy as np
from dotenv import load_dotenv
import requests
import json
from datetime import datetime
from panel import widgets

import seaborn as sns
%matplotlib inline
```

```python
file_path2 = Path("folder/subfolder/df_Merged.csv")
df_merged = pd.read_csv(file_path2, index_col="DATE")


file_path = Path("folder/subfolder/Percentage_change.csv")
df_percentchange = pd.read_csv(file_path, index_col="DATE")

file_path3 = Path("folder/subfolder/mcsimulation.csv")
mcsimulation = pd.read_csv(file_path3)

correlation = df_merged.corr()
```

```python
def currency_1():
    
    Percentage_change= df_percentchange.hvplot(title="Cryptocurrencies Percent Change Since 05/10/19", figsize=(18, 15))
 
    return Percentage_change

def currency_2():
    
    Percentage_change_Correlation=df_percentchange.hvplot(kind='scatter', x='ETH', y='BTC', title="Percent Change Correlation - BTC vs. ETH")
    
    return Percentage_change_Correlation

def currency_3():
    
    Eth_Btc_PriceChange = df_merged.hvplot(kind='scatter', x='ETH', y='BTC', title="Price Change Correlation - BTC vs ETH")

    return Eth_Btc_PriceChange 

def currency_5():
     
    correlation = df_merged.corr()
    heat_map = correlation.hvplot.heatmap(x=[], y= [], C='Correlation', height=500, width=500, colorbar=False)
    
    
    return heat_map

def currency_6():
    
    prices = df_merged.hvplot.area(figsize = (12,12), subplots=True, title="Crypto Prices",shared_axes=False)

    return prices

def currency_7():
    
    # Plot a rolling 7-day mean of Crypto's closing prices
    seven_daymean = df_merged.rolling(window=7).mean().hvplot(figsize = (12,12), subplots=True,shared_axes=False)

    return seven_daymean

def currency_8():
    
    # Plot a rolling 7-day std dev of Crypto's closing prices
    seven_day_Deviation = df_merged.rolling(window=7).std().hvplot(figsize = (12,12), subplots=True,shared_axes=False)

    return seven_day_Deviation

def currency_9():
    
    # Plot a rolling 30-day mean of Crypto's closing prices
    Thirty_mean_Closing =df_merged.rolling(window=30).mean().hvplot(figsize = (12,12), subplots=True,shared_axes=False)

    return Thirty_mean_Closing

def currency_10():
    
    # Plot a rolling 30-day std dev of Crypto's closing prices
    Thirty_STD_Closing = df_merged.rolling(window=30).std().hvplot(figsize = (12,12), subplots=True,shared_axes=False)

    return Thirty_STD_Closing

def currency_11():
    
    # Plot a rolling 180-day mean of Crypto's closing prices
    Three_Month_mean= df_merged.rolling(window=180).mean().hvplot(figsize = (12,12), subplots=True,shared_axes=False)

    return Three_Month_mean

def currency_12():
    
    # Plot a rolling 180-day std dev of Crypto's closing prices
    Three_Month_STD= df_merged.rolling(window=180).std().hvplot(figsize = (12,12), subplots=True,shared_axes=False)

    return Three_Month_STD

def currency_18():
    
    McSim = mcsimulation.hvplot(legend=False, figsize=(15,15))
    
    return McSim

def currency_19():
    ax = mcsimulation.hvplot.hist( figsize=(10, 8), legend=False)
    
    return ax
```

```python
# Create a Title for the Dashboard

dashboard_title = "Cryptocurrency Portfolio Analysis"

Info = pn.pane.Markdown(
""""
    
    Info ,
    1- Historical Price Action,
    2- Deviation,
    3- Futures and Forecasts,
    4- Closing
   """
)


tab_1 = pn.Column(
    ("#Closing Prices over numerous time periods"),
    ("##Including Moving Averages"),
    currency_6(),
    ("##Seven Day Rolling Average"),
    currency_7(),
    ("##Thirty Day Rolling Average"),
    currency_9(),
    ("##180 Day Rolling Average"),
    currency_11()
)

tab_2 = pn.Column(
    ("#Deviation and Correlation"),
    currency_1(),
    ("##Percent Change Correlation - BTC vs. ETH"),
    currency_2(),
    ("##Price Change Correlation - BTC vs ETH"),
    currency_3(),
    ("##Corollary HeatMap"),
    currency_5(),
    ("##Seven Day STD Deviation"),
    currency_8(),
    ("##Thirty Day STD Deviation"),
    currency_10(),
    ("##180 Day Rolling STD Deviation"),
    currency_12()
    
)


tab_3 = pn.Column(
    ("#Statistical Possibilities"),
    currency_18(),
    currency_19()
)

tab_4 = pn.Column(
    ("#About")
)

    
# Create the dashboard

panel = pn.Tabs(
        ("Intro",Info),
        ("Historical Price Action", tab_1),
        ("Deviation", tab_2),
        ("Futures and Forecasts", tab_3),
        ("About", tab_4)    
)
```

```python
# Serve the dashboard
print(dashboard_title)
panel.servable()
```
